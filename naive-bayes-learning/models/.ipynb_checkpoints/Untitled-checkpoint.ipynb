```python
import matplotlib.pyplot as plt
import numpy as np

# data to plot
n_groups = 3
size_list = ["5000", "10000", "15000"]
clone_list = [0.058, 0.051, 0.078]
mmap_list = [0.0018, 0.0018, 0.0018]
write_list = [0.004, 0.007, 0.013]
open_list = [0.0092, 0.007, 0.0054]

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.2
opacity = 0.8

rects1 = plt.bar(index, clone_list, bar_width,
alpha=opacity,
color='b',
label='clone')

rects2 = plt.bar(index + bar_width, mmap_list, bar_width,
alpha=opacity,
color='g',
label='mmap')

rects2 = plt.bar(index + 2*bar_width, write_list, bar_width,
alpha=opacity,
color='r',
label='write')

rects2 = plt.bar(index + 3*bar_width, open_list, bar_width,
alpha=opacity,
color='y',
label='open')

plt.xlabel('File size')
plt.ylabel('time (ms)')
plt.title('Systemcalls Function Compare Graph')
plt.xticks(index + bar_width, size_list)
plt.legend()

plt.tight_layout()
plt.show()
```
